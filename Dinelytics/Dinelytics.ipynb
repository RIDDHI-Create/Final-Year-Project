```python
import pandas as pd
dataframe = pd.read_csv("Zomato data.csv",encoding="ISO-8859-1") 
def handleRate(value):
    value = str(value).split('/')
    value = value[0];
    return float(value)
dataframe['rate']=dataframe['rate'].apply(handleRate)
print(dataframe)
dataframe.info()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
dataframe = pd.read_csv("Zomato data.csv",encoding="ISO-8859-1") 
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': grouped_data})
plt.plot(result, c="green",marker="o")
plt.xlabel("Types of restaurant",c="red", size=20)
plt.ylabel("Votes", c="red",size=20)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
dataframe = pd.read_csv("Zomato data.csv",encoding="ISO-8859-1") 
pivot_table = dataframe.pivot_table(index = "listed_in(type)",columns = "online_order", aggfunc="count", fill_value=0)
plt.figure(figsize=(10,6))
sns.heatmap(pivot_table , annot = True , cmap = "coolwarm", fmt = "d")
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed In (Type)")
plt.show()
```

```python
# Conclusion :- Dinning restaurant primarily accept offline orders,
                # whereas cafes primarily recevies online order.
                # this suggest that clients prefer to place orders in person at 
               # restaurant , but prefer online ordering at cafes.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
# Change to your actual file path
df = pd.read_csv("Zomato data.csv",encoding="ISO-8859-1")

# Convert 'rate' column to numeric (removing '/5' and handling missing values)
df['rate'] = df['rate'].str.replace('/5', '', regex=True).astype(float)

# Create a figure with subplots
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# 1. Countplot of Restaurant Types
sns.countplot(y=df['listed_in(type)'], order=df['listed_in(type)'].value_counts().index, ax=axes[0, 0])
axes[0, 0].set_title("Distribution of Restaurant Types")

# 2. Online Order Availability
sns.countplot(x=df['online_order'], ax=axes[0, 1])
axes[0, 1].set_title("Online Order Availability")

# 3. Table Booking Availability
sns.countplot(x=df['book_table'], ax=axes[1, 0])
axes[1, 0].set_title("Table Booking Availability")

# 4. Ratings Distribution
sns.histplot(df['rate'].dropna(), bins=10, kde=True, ax=axes[1, 1])
axes[1, 1].set_title("Restaurant Ratings Distribution")

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# Conclusion:- Most Restaurants are categorized under "Dinning". with limited online ordering and 
# table booking options.Rating mostly range between 3.5 & 4.2. ,Indicting good but not exceptional
# service.Expanding online services and improving quality can enhance customer experience and ratings.
```
